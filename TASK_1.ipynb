```python
import pandas as pd
data = pd.read_csv("task pop.csv")
import matplotlib.pyplot as plt
```

```python
# GROUPING DATA TOGETHER AND COUNTING OCCURENCES

country_population = data.groupby(['Country Name','2022']).size().unstack(fill_value=0)
```

```python




# PLOTTING BAR CHART
country_population.plot(kind='bar',figsize=(10,6))
plt.xlabel("countries")
plt.ylabel("Count")
plt.title("population of countries")
plt.legend(title='population',labels = country_population.columns)
plt.show()
```
